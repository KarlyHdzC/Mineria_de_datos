`describe()` en pandas es una herramienta extremadamente Ãºtil para obtener un resumen estadÃ­stico de los datos de un DataFrame o Serie.

### Â¿QuÃ© hace `describe()`?
Proporciona estadÃ­sticas descriptivas que resumen la tendencia central, la dispersiÃ³n y la forma de la distribuciÃ³n de los datos de un DataFrame o Serie.

Por defecto, solo funciona con **columnas numÃ©ricas**, aunque tambiÃ©n puedes aplicarlo a datos categÃ³ricos.

---

### Â¿QuÃ© incluye el resumen?
Para datos **numÃ©ricos**, incluye:
1. **count**: NÃºmero de valores no nulos.
2. **mean**: Media aritmÃ©tica.
3. **std**: DesviaciÃ³n estÃ¡ndar.
4. **min**: Valor mÃ­nimo.
5. **25%**, **50% (mediana)**, **75%**: Cuartiles.
6. **max**: Valor mÃ¡ximo.

Para datos **categÃ³ricos** (`include='object'`), incluye:
1. **count**: NÃºmero de valores no nulos.
2. **unique**: NÃºmero de valores Ãºnicos.
3. **top**: Valor mÃ¡s frecuente (moda).
4. **freq**: Frecuencia del valor mÃ¡s frecuente.

```python
new_laptops_df.describe()
```

```python
new_laptops_df.describe(include="object")
```

```python
# Calcular el precio promedio por marca
average_prices_by_brand = new_laptops_df.groupby('brand')['price'].mean().sort_values()

# Mostrar las marcas con los precios promedio mÃ¡s bajos (por ejemplo, las 5 mÃ¡s baratas)
display(average_prices_by_brand.head())
```

```python
# Seleccionar las marcas mÃ¡s baratas (por ejemplo, las 5 primeras de la lista ordenada)
cheapest_brands = average_prices_by_brand.head(5).index.tolist()

# Filtrar el DataFrame original para incluir solo estas marcas
cheapest_laptops_df = new_laptops_df[new_laptops_df['brand'].isin(cheapest_brands)]

display(cheapest_laptops_df.head())
```

```python
# Cuantos valores Ãºnicos tiene breadCrumbs

new_laptops_df["brand"].unique()
```

Uso de min() y max()
1. min():
* Encuentra el valor mÃ¡s pequeÃ±o en los datos.
2. max():
* Encuentra el valor mÃ¡s grande en los datos.
Ambos mÃ©todos pueden aplicarse a datos numÃ©ricos y categÃ³ricos.

```python
# Operamos sobre max y min de descuento

min_discount = new_laptops_df["discount_percentage"].min()
max_discount = new_laptops_df["discount_percentage"].max()

print("Descuento mÃ­nimo", min_discount)
print("Descuento mÃ¡ximo:", max_discount)
```
