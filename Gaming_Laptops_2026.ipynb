`describe()` en pandas es una herramienta extremadamente útil para obtener un resumen estadístico de los datos de un DataFrame o Serie.

### ¿Qué hace `describe()`?
Proporciona estadísticas descriptivas que resumen la tendencia central, la dispersión y la forma de la distribución de los datos de un DataFrame o Serie.

Por defecto, solo funciona con **columnas numéricas**, aunque también puedes aplicarlo a datos categóricos.

---

### ¿Qué incluye el resumen?
Para datos **numéricos**, incluye:
1. **count**: Número de valores no nulos.
2. **mean**: Media aritmética.
3. **std**: Desviación estándar.
4. **min**: Valor mínimo.
5. **25%**, **50% (mediana)**, **75%**: Cuartiles.
6. **max**: Valor máximo.

Para datos **categóricos** (`include='object'`), incluye:
1. **count**: Número de valores no nulos.
2. **unique**: Número de valores únicos.
3. **top**: Valor más frecuente (moda).
4. **freq**: Frecuencia del valor más frecuente.

```python
laptops_df.describe()
```

```python
laptops_df.describe(include="object")
```

```python
# Calcular el precio promedio por marca
average_prices_by_brand = laptops_df.groupby('brand')['price'].mean().sort_values()

# Mostrar las marcas con los precios promedio más bajos (por ejemplo, las 5 más baratas)
display(average_prices_by_brand.head())
```

```python
# Seleccionar las marcas más baratas (por ejemplo, las 5 primeras de la lista ordenada)
cheapest_brands = average_prices_by_brand.head(5).index.tolist()

# Filtrar el DataFrame original para incluir solo estas marcas
cheapest_laptops_df = laptops_df[laptops_df['brand'].isin(cheapest_brands)]

display(cheapest_laptops_df.head())
```

Uso de min() y max()
1. min():
* Encuentra el valor más pequeño en los datos.
2. max():
* Encuentra el valor más grande en los datos.
Ambos métodos pueden aplicarse a datos numéricos y categóricos.

```python
# Operamos sobre max y min de descuento

min_discount = laptops_df["discount_pct"].min()
max_discount = laptops_df["discount_pct"].max()

print("Descuento mínimo", min_discount)
print("Descuento máximo:", max_discount)
```
