Construye un diccionario con los dos discursos como hicimos en clase con los niveles:

Con ese diccionario responde:

¿Qué orador menciona más veces la palabra "failure"? ¿Y "success"? Rowling los dos

¿Cuáles son los 10 trigramas más frecuentes de cada discurso?
En la ejecución

Genera una gráfica de barras comparando las 5 palabras más frecuentes por orador, excluyendo stopwords ("the", "and", "to", "of", "a", "i", "in", "is", "it", "you").

```python

# Diccionario de discursos
discursos = {
    "Jobs": contenido_1,
    "Rowling": contenido_2
}

stopwords = {"the", "and", "to", "of", "a", "i", "in", "is", "it", "you", "s", "he", "his", "that","with","was","from", "this", "by"}

def limpiar_texto(texto):
    texto = texto.lower()
    palabras = re.findall(r'\b[a-z]+\b', texto)
    return palabras

conteo_palabras = {}

for orador, texto in discursos.items():
    palabras = limpiar_texto(texto)
    conteo = Counter(palabras)
    conteo_palabras[orador] = conteo

print("Frecuencia de 'failure':")
for orador in discursos:
    print(orador, conteo_palabras[orador]["failure"])

print("\nFrecuencia de 'success':")
for orador in discursos:
    print(orador, conteo_palabras[orador]["success"])

print("\nQuien menciona más 'failure':",
      max(discursos, key=lambda x: conteo_palabras[x]["failure"]))

print("Quien menciona más 'success':",
      max(discursos, key=lambda x: conteo_palabras[x]["success"]))

def trigramas(palabras):
    return zip(palabras, palabras[1:], palabras[2:])

print("\nTop 10 trigramas por discurso:\n")

for orador, texto in discursos.items():
    palabras = limpiar_texto(texto)
    tri = Counter(trigramas(palabras))

    print(orador)
    for trig, freq in tri.most_common(10):
        print(trig, freq)
    print()


top_palabras = {}

for orador, texto in discursos.items():
    palabras = limpiar_texto(texto)
    palabras_filtradas = [p for p in palabras if p not in stopwords]
    conteo = Counter(palabras_filtradas)
    top_palabras[orador] = conteo.most_common(5)


# --- Gráfica de barras ---
plt.figure(figsize=(10,6))

for i, (orador, datos) in enumerate(top_palabras.items()):
    palabras = [p[0] for p in datos]
    freqs = [p[1] for p in datos]

    posiciones = [x + i*0.4 for x in range(len(palabras))]
    plt.bar(posiciones, freqs, width=0.4, label=orador)

plt.xticks(range(5), [p[0] for p in top_palabras[list(top_palabras.keys())[0]]])
plt.xlabel("Palabras")
plt.ylabel("Frecuencia")
plt.title("Top 5 palabras más frecuentes por orador (sin stopwords)")
plt.legend()

plt.show()
```
